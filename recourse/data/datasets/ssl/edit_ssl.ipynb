```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("chicago-ssl-clean.csv")
```

```python
age_d = {'less than 20': (14, 20), '20-30': (20, 30), '30-40': (30, 40), '40-50': (40, 50), '50-60': (50, 60), '60-70': (60, 70), '70-80': (70, 80) }
data['PREDICTOR RAT AGE AT LATEST ARREST'] = data['PREDICTOR RAT AGE AT LATEST ARREST'].apply(lambda x: np.random.randint(*age_d[x]))
```

```python
data
```

```python
data['SSL SCORE'] = data['SSL SCORE'].apply(lambda x: 0 if x > 345 else 1)
```

```python
data.drop(columns=['PREDICTOR RAT GANG AFFILIATION', 'SEX CODE CD'], inplace=True)
```

```python
data['RACE CODE CD'] = data['RACE CODE CD'].apply(lambda x: 'w' if x in ['WHI', 'WWH'] else 'nw')
```

```python
data
```

```python
new_cols = list(data.columns)[1:] + ['SSL SCORE']
```

```python
data = data[new_cols]
```

```python
data
```

```python
data.to_csv("ssl_numeric.csv", index=False)
```
